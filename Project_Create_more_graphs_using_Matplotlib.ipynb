Letâ€™s imagine that you are working as a high school math teacher. You want to display important metrics about your current class and the ones you have taught in previous years, in order to understand patterns and trends that may not be visible from just seeing the numbers in a spreadsheet. In this project, we will create some graphs.

```python
from matplotlib import pyplot as plt
import numpy as np
```

**Bar Chart with Error**

```python
past_years_averages = [82, 84, 83, 86, 74, 84, 90]
years = [2000, 2001, 2002, 2003, 2004, 2005, 2006]
error = [1.5, 2.1, 1.2, 3.2, 2.3, 1.7, 2.4]
```

```python
# create a figure (10,8)
plt.figure(figsize=(10,8))
# plot bars and errors
plt.bar(range(len(past_years_averages)), past_years_averages, yerr=error, capsize=5) # yerr: error +- , capsize: size of caps, i.e. the horizontal line at errors (-)
# set the specific range of values on x ang y axis
plt.axis([-0.5,6.5,70,95])
# x-ticks and labels
ax = plt.subplot()
ax.set_xticks(range(len(years)))
ax.set_xticklabels(years)
# add titles to the graph
plt.title('Final Exam Averages')
plt.xlabel('Year')
plt.ylabel('Test average')
# save the figure
plt.savefig('my_bar_chart.png')
# plot the graph
plt.show()
```

**Side By Side Bars**

```python
# Units of Maths
unit_topics = ['Limits', 'Derivatives', 'Integrals', 'Diff Eq', 'Applications']

# Average test scores from school A
middle_school_a = [80, 85, 84, 83, 86]

# Average test scores from school B
middle_school_b = [73, 78, 77, 82, 86]

def create_x(t, w, n, d):
    '''
    Calculate x-coordinates for side-by-side bars.

    Parameters:
    t : number of total datasets
    w : width of each bar
    n : number of current dataset
    d : number of sets of bars (e.g., 5 in this case)

    Returns:
    A list of x-coordinates for each set of bars.
    '''
    return [t*x + w*n for x in range(d)]

# Creation of bars for school A and B
school_a_x = create_x(2, 0.8, 1, 5)
school_b_x = create_x(2, 0.8, 2, 5)

# Create figure (10,8)
plt.figure(figsize=(10,8))

# Define axes
ax = plt.subplot()

# Plot side-by-side bars for each school
plt.bar(school_a_x, middle_school_a)
plt.bar(school_b_x, middle_school_b)

# Calculate middle x positions for x ticks
middle_x = [(a + b) / 2.0 for a, b in zip(school_a_x, school_b_x)]

# Set x ticks and labels
ax.set_xticks(middle_x)
ax.set_xticklabels(unit_topics)

# Add legend, x and y labels, and title
plt.legend(['Middle School A', 'Middle School B'], loc=2)
plt.xlabel('Unit')
plt.ylabel('Test Average')
plt.title('Test Averages on Different Units')

# Save figure
plt.savefig('my_side_by_side.png')

plt.show()
```

**Stacked Bars**

```python
# Define the unit topics
unit_topics = ['Limits', 'Derivatives', 'Integrals', 'Diff Eq', 'Applications']

# Number of students who received each grade level for each unit
As = [6, 3, 4, 3, 5]
Bs = [8, 12, 8, 9, 10]
Cs = [13, 12, 15, 13, 14]
Ds = [2, 3, 3, 2, 1]
Fs = [1, 0, 0, 3, 0]

# Define the x values as the range of the number of units
x = range(5)

# Calculate the bottom values for each grade level to stack the bars properly
c_bottom = np.add(As, Bs)
d_bottom = np.add(c_bottom, Cs)
f_bottom = np.add(d_bottom, Ds)

# Create a figure for the plot
plt.figure(figsize=(10,8))

# Plot stacked bars for each grade level
plt.bar(x, As)
plt.bar(x, Bs, bottom=As)
plt.bar(x, Cs, bottom=c_bottom)
plt.bar(x, Ds, bottom=d_bottom)
plt.bar(x, Fs, bottom=f_bottom)

# Define axes
ax = plt.subplot()

# Set x ticks and labels
ax.set_xticks(range(len(unit_topics)))
ax.set_xticklabels(unit_topics)

# Add legend, labels and title
plt.legend(['A', 'B', 'C', 'D', 'F'], loc=3)
plt.xlabel('Unit')
plt.ylabel('Number of students')
plt.title('Grade distribution')

# Save the figure
plt.savefig('my_stacked_bar.png')

# Display the plot
plt.show()
```

**Labeled Pie Chart**

```python
# Define the unit topics and the number of times each topic was reported as the hardest
unit_topics = ['Limits', 'Derivatives', 'Integrals', 'Diff Eq', 'Applications']
num_hardest_reported = [1, 3, 10, 15, 1]

# Create a new figure for the plot
plt.figure(figsize=(10,8))

# Plot a pie chart with the number of times each topic was reported as the hardest, labeled with unit topics and percentage of unit difficulty (as reported)
plt.pie(num_hardest_reported, labels=unit_topics, autopct='%d%%')

# Set the aspect ratio to be equal to ensure a circular pie chart
plt.axis('equal')

# Add title to the plot
plt.title('Hardest Topics')

# Save the figure
plt.savefig('my_pie_chart.png')

# Display the plot
plt.show()
```

**Line with Shaded Error**

```python
# Define the self-reported hours spent studying and the corresponding exam scores
hours_reported = [3, 2.5, 2.75, 2.5, 2.75, 3.0, 3.5, 3.25, 3.25, 3.5, 3.5, 3.75, 3.75, 4, 4.0, 3.75, 4.0, 4.25, 4.25, 4.5, 4.5, 5.0, 5.25, 5, 5.25, 5.5, 5.5, 5.75, 5.25, 4.75]
exam_scores = [52.53, 59.05, 61.15, 61.72, 62.58, 62.98, 64.99, 67.63, 68.52, 70.29, 71.33, 72.15, 72.67, 73.85, 74.44, 75.62, 76.81, 77.82, 78.16, 78.94, 79.08, 80.31, 80.77, 81.37, 85.13, 85.38, 89.34, 90.75, 97.24, 98.31]

# Create a new figure for the plot
plt.figure(figsize=(10,8))

# Plot the line (width = 0.2) graph of exam scores vs. hours studied
plt.plot(exam_scores, hours_reported, linewidth=2)

# Calculate the lower and upper bounds for the shaded region
hours_lower_bound = [hour - 0.2*hour for hour in hours_reported]
hours_upper_bound = [hour + 0.2*hour for hour in hours_reported]

# Shade the area between the lower and upper bounds
# alpha: the shaded error
plt.fill_between(exam_scores, hours_lower_bound, hours_upper_bound, alpha=0.2)

# Add title and labels to the plot
plt.title('Time spent studying vs final exam scores')
plt.xlabel('Score')
plt.ylabel('Hours studying (self-reported)')

# Save the figure
plt.savefig('my_line_graph.png')

# Display the plot
plt.show()
```
